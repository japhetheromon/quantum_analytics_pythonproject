#### Working with the Valuation column
- Removing the '$' sign
- Rename the Valuation column header
- change the STR to FLOAT
- Convert each value to billions

```python
# Renaming the Column name 
df = df.rename(columns={'Valuation' : 'Valuation ($B)'})

# Removing the '$' sign from the Valuation column
df['Valuation ($B)'] = df['Valuation ($B)'].replace('$', '', regex=False)

# Removing the 'B' sign from the Valuation column and converting from STR to FLOAT
df['Valuation ($B)'] = df['Valuation ($B)'].replace('B', '', regex=False).astype('float')
```

```python
df['Valuation ($B)'].describe()
```

```python
df.head()
```

#### Working with the "Funding" Column
- Remove the '$' sign
- Convert to billion
- Convert from STR to INT
- Resolve the columns where funding is "Unknown"

```python
df['Funding'].value_counts()
```

##### The data in the Funding column have '$', 'M', 'B', 'K' symbols which is need to be replaced and making million as an standard for measuring the amount of Funding

```python
# Renaming the Column name 
df = df.rename(columns={'Funding' : 'Funding ($M)'})

# Removing the '$' sign from the Funding column
df['Funding ($M)'] = df['Funding ($M)'].str.replace('$', '', regex=False)

# Removing the 'M' sign from the Funding column 
df['Funding ($M)'] = df['Funding ($M)'].str.replace('M', '', regex=False)

# Replacing 'Unknown' with 0
df['Funding ($M)'] = df['Funding ($M)'].str.replace('Unknown', '0', regex=False)

# Replacing 'B' sign from the funding column and  converting from STR to FLOAT
df['Funding ($M)'] = df['Funding ($M)'].str.replace('B', '', regex=False).astype('float').apply(lambda x: x*1000)
```

```python
# Checking
df['Funding ($M)'].describe()
```

#### Spliting Select Investor Column

```python
#splitting into new columns
df[['Investor 1', 'Investor 2', 'Investor 3', 'Investor 4']] = df['Select Investors'].str.split(', ', expand=True) 

#dropping column
df.drop(columns='Select Investors', inplace=True) 
```

```python
# Checking
df.head()
```

#### Changing Date Joined to Datetime format

```python
# Setting date to datetime
df['Date Joined'] = pd.to_datetime(df['Date Joined'], format='%m/%d/%Y') 

# Creating "Year Joined" Column
df['Year Joined'] = df['Date Joined'].dt.year
```

```python
df.head()
```

```python
# Creating 'Return On Investment' Column
df['Return on Investment ($M)'] = (df['Valuation ($B)']*1e9 - df['Funding ($M)']*1e6)/1e6
df.head()

```

# Data Visualization

### Which unicorn companies have had the biggest return on investment?

```python
company_roi = df['Return on Investment ($M)'].groupby(df.Company).sum().sort_values(ascending=False)[:10]
company_roi
```

```python
# visualize using bar chart
ax = company_roi.plot(kind = 'bar', figsize = (10,5), title = 'Top 10 Companies with the biggest Return on Investment ($M)', xlabel = 'Company',
                       ylabel = 'Return on Investment', legend = False)
#annotate
ax.bar_label(ax.containers[0], label_type = 'edge')

# set x-axis tick labels and rotation
ax.set_xticklabels(company_roi.index, rotation=45, ha='right')

# pad the spacing between the number and edge of teh figure
ax.margins(y = 0.1)

# show the visual
plt.show()
```

#### Observation
- It is observed from the bar chart that Bytedance has the highest Return on Investment with a total of 172 billion dollars followed by SHEN with 98 billion dollars while the least top 10 companies is FTX with a Return on Investment of 30 billion dollars

### How long does it usually take for a company to become a unicorn? Has it always been this way?

```python
# Create a column showing the number of years as a Unicorn
df['Unicorn Years'] = df['Year Joined'] - df['Year Founded']
df['Unicorn Years']
```

```python
df['Unicorn Years'].describe()
```

```python
# Creating a Scatter plot chart to visualize the average number of years
df.plot.scatter(x = 'Year Joined', y = 'Unicorn Years', figsize=(8,10))

# label the axis
plt.xlabel('Year Joined')
plt.ylabel('No. of Years as a Unicorn')

# set the title
plt.title('Average Numbers of Years to become a Unicorn')


# Adding a horizontal line at mean value of y
plt.axhline(y=np.nanmean(df['Unicorn Years']))

# show the visual
plt.show()
```

#### Observation
